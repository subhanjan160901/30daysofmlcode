```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
flat_chess = cv2.imread("C:/Users/dell/Documents/Udemy OpenCV course/flat_chessboard.png")
```

```python
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
real_chess = cv2.imread("C:/Users/dell/Documents/Udemy OpenCV course/real_chessboard.jpg")
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray_flat_chess
```

**We will need float point values for harris-corner detection algorithm.**

```python
gray = np.float32(gray_flat_chess)
gray
```

```python
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01*dst.max()] = [255,0,0]  #RGB
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
gray
```

```python
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
real_chess[dst>0.01*dst.max()] = [255,0,0]  #RGB
```

```python
plt.imshow(real_chess)
```
